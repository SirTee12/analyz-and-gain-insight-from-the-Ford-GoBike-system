## Investigation Overview
> The goal of the investigation is to analyze the Ford gobike system and determine the category of people that uses the system more, the duration and distance of ride as well as the effect of gender on them

> These are the main insights:
+ Men are the majority users of the ford gobike system and most of them are young people
+ Gender is not a factor in determining duration and time of travel ( boxplots will be used instead of violin plot)

## Dataset Overview

> This data set includes information about individual rides made in a bike-sharing system covering the greater San Francisco Bay area. This dataset was gottrn from one of Udacity's curated datasets. It consists of 16 columns with 180k+ records. The Ford GoBike system dataset provides anonymized, timestamped data about the start and end station for a bike, the duration of the ride, the user type (subscriber or customer), some customer-reported attributes like birth year and gender (the gender here is divide into three categories: male, female, other), as well as the the start and end time for each ride.

```python
# import all packages and set plots to be embedded inline
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

%matplotlib inline

# suppress warnings from final output
import warnings
warnings.simplefilter("ignore")
```

```python
# load in the dataset into a pandas dataframe
ford_bike = pd.read_csv('ford_go_df.csv', parse_dates=['start_time', 'end_time'])
```

## INSIGHTS

> There are three genders using the ford gobike system (male, female, other) and the men are the majority users of the system accounting for more than 130k members

```python
# set the graph style
sns.set_style('darkgrid')
# set the figure size
plt.rcParams['figure.figsize'] = (12, 10)
```

```python
# set order of bins
bin_order = ford_bike['member_gender'].value_counts().index

# draw countplot
v1 = sns.countplot(data = ford_bike, x = 'member_gender', order = bin_order, color = sns.color_palette()[3])

# set tick labels, axis labels and title
v1.set_yticklabels(['0', '20k', '40k', '60k', '80k', '100k', '120k' ], fontsize = 13);
v1.set_xticklabels(['Male', 'Female', 'Other'], fontsize = 13);
v1.set_xlabel('Gender', fontsize = 15)
v1.set_ylabel('Number of users', fontsize = 15)
v1.set_title('User population of ford gobike system based on gender', fontsize = 20);
```

> There are about 175000+ users and majority of them are young people between the ages 16 - 39 year accounting for more than 90% of the total users. Only very few people in their 60s use the system

```python
# set bin order
bin_order_1 = ford_bike['age_group'].value_counts().index
# draw a count plot
v2 = sns.countplot(data = ford_bike, x = 'age_group', order = bin_order_1, color = sns.color_palette()[3]);

# set axis, tick and title labels
v2.set_yticklabels(['0', '20k', '40k', '60k', '80k', '100k', '120k', '140k', '160k' ], fontsize = 13)
v2.set_xticklabels(['16-39', '40-60', '60+'], fontsize = 13)
v2.set_xlabel('Age Group', fontsize = 15)
v2.set_ylabel('Number of users', fontsize = 15)
v2.set_title('User population of ford gobike system based for each age group', fontsize = 20);
```

> Apart from the fact that the population of men using the system more than triples the population of women and is about 10 times those of gender **other**, one would expect men to ride for longer duration as compared to women or the other group due to the difference in physical strengths but the situation here is different as women ride for longer duration than men although the difference is not much

> Majority of the users ride for less than 20 minutes across all genders which is 75% of the population for each gender.

```python
# draw a box plot
# subset data for duration less than 40 minutes since more than 90% of the users didn't ride for more than this
data_range = ford_bike[ford_bike['duration_min'] < 40]

# draw boxplot
v3 = sns.boxplot(data = data_range, x = 'member_gender',
                 y = 'duration_min', color = sns.color_palette()[3], 
                sym = '')

# set axis, tick and title labels
v3.set_xticklabels(['Male', 'Female', 'Other'], fontsize = 13)
plt.xlabel('Gender', fontsize = 15)
plt.ylabel('Duration of ride (minutes)', fontsize = 15)
v3.set_title('Ditsribution of duration across each gender', fontsize = 20, y = 1.05);
```

> one would expect men to ride for longer distance as compared to women or the other group due to the difference in physical strengths but the situation here is different as women ride for longer distance than men and closely followed by the other gender although the difference is not much. 

> Majority of the users ride for less than 4km across all gender and only a very few ride above 4km 

```python
# subset data for users with distance less than 10km as most of the users are observed to ride for less than this
data_range_2 = ford_bike[ford_bike['distance_km'] < 10]

# Draw boxplot
v4 = sns.boxplot(data = data_range_2, x = 'member_gender',
                 y = 'distance_km', color= sns.color_palette()[3],
                sym = '')

# set axis, tick and title labels
v4.set_xticklabels(['Male', 'Female', 'Other'], fontsize = 13);
v4.set_xlabel('Gender', fontsize = 15)
v4.set_ylabel('Distance (Km)', fontsize = 15)
v4.set_title('Ditsribution of distance across each gender', fontsize = 20, y = 1.05);
```
